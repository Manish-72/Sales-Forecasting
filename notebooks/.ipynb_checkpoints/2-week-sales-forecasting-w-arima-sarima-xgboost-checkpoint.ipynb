# 2 Week Sales Forecasting w/ ARIMA / SARIMA / XGBoost
## Objective
- Evaluate ARIMA, SARIMA, and XGBoost model performance for a 2 week daily sales forecast
## Results
- ARIMA - auto and manually tuned parameters, poor performance MAPE > 100% , model could not capture historical data trend
- SARIMA -  auto and manually tuned parameters, poor performance MAPE > 100% , model could not capture historical data trend
- XGBoost - hyperparamter optimization with grid search, leveraged additional numerical features (sales lag, rolling averages), better performance than ARIMA/SARIMA but MAPE > 70% so model is not very useful


The dataset exhibits sales records over 4 years: between January 2015 and December 2018. It contains 18 columns (features) and 9,800 rows (instances).
The columns are described as follows:
* Row ID: An integer value representing a unique sequence ID for the rows in the dataset.
* Order ID: An alphanumeric identifier for the purchase order. It can appear multiple times if the order included multiple items.
* Order Date: The date when the purchase order was placed.
* Ship Date: The date when the order was shipped out from the retailer.
* Ship Mode: A string indicating the priority level of the transportation (e.g., First Class, Standard Class).
* Customer ID: An alphanumeric customer identifier that appears to be anonymized.
* Customer Name: A string showing the name of the customer, which also appears to be anonymized.
* Segment: A string describing the kind of customer: Consumer, Corporate, or Home Office.
* Country: A string indicating that the dataset only includes orders from the United States.
* City: A string representing the US city from which the order originated.
* State: A string representing the US state from which the order originated.
* Post Code: An alphanumeric value that appears to be the zip code.
* Region: A string indicating a high-level geographical division of the US (e.g., East, West).
* Product ID: An alphanumeric product identifier.
* Category: A string describing the type of product: Furniture, Office Supplies, or Technology.
* Sub-Category: A string for the sub-type of the product (e.g., Chairs, Phones, Envelopes).
* Product Name: A string showing the name of the item sold.
* Sales: A float representing the monetary amount of the sold products (assumed to be in USD).

```python
# Drop irrevelant columns that are just IDs
df_cleaned = df.drop(columns=['Row ID', 'Order ID', 'Customer Name', 'Product ID', 'Product Name'])

# Convert Order Date and Ship Date from object to datetime
df_cleaned['Order Date'] = pd.to_datetime(df_cleaned['Order Date'], format="%d/%m/%Y")
df_cleaned['Ship Date'] = pd.to_datetime(df_cleaned['Ship Date'], format="%d/%m/%Y")

# Extract useful features from the date fields
df_cleaned['Order Year'] = df_cleaned['Order Date'].dt.year
df_cleaned['Order Month'] = df_cleaned['Order Date'].dt.month
df_cleaned['Order Day'] = df_cleaned['Order Date'].dt.day
# shipping delay
df_cleaned['Ship Day Diff'] = (df_cleaned['Ship Date'] - df_cleaned['Order Date']).dt.days
# Convert 'Order Date' to a numerical format (e.g., days since start of dataset)
df_cleaned['Days Since Start'] = (df_cleaned['Order Date'] - df_cleaned['Order Date'].min()).dt.days

# Encode categorical columns
# One hot encode categorical variables (low carinality)
df_cleaned = pd.get_dummies(df_cleaned, columns=['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category'])

# Label encode those with high cardinality
from sklearn.preprocessing import LabelEncoder

label_enc_cols = ['Country', 'State', 'City']
encoder = LabelEncoder()

for col in label_enc_cols:
    df_cleaned[col] = encoder.fit_transform(df_cleaned[col])

print(df_cleaned.describe())
df_cleaned.drop(columns=['Customer ID', 'Country'], inplace=True)
```

## ARIMA and SARIMA
* Used Auto ARIMA and Auto SARIMA to find optimal parameters but when using them, results were not as good as manual
* Used the Augmented Dickey-Fuller (ADF) test as a statistical test to confirm this time series data is stationary 

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Step 1: Load & Preprocess Data
# Load dataset from CSV
df = pd.read_csv("../Dataset.csv")

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")

# Aggregate sales to daily total (in case multiple orders occur on the same day)
df = df.groupby('Order Date')['Sales'].sum().reset_index()

# Create a complete date range to fill missing days with 0 sales
date_range = pd.date_range(df['Order Date'].min(), df['Order Date'].max(), freq='D')
df_full = pd.DataFrame({'Order Date': date_range}).merge(df, on='Order Date', how='left').fillna({'Sales': 0})

# Set 'Order Date' as index
df_full.set_index('Order Date', inplace=True)

# Ensure the index has a daily frequency (fix for newer Pandas versions)
df_full = df_full.resample('D').asfreq().fillna(0)  # Resample and fill missing days with 0
```

```python
import pmdarima as pm

# Step 2: Train ARIMA & SARIMA on Past Data
# Define training data (excluding the last 14 days for testing)
train_data = df_full.iloc[:-14]
test_data = df_full.iloc[-14:]

"""
# Use Auto ARIMA to find the best (p,d,q) for ARIMA
arima_model = pm.auto_arima(train_data['Sales'], seasonal=False, 
                            trace=True, suppress_warnings=True, stepwise=True)

# Print best parameters
print("Best ARIMA Parameters:", arima_model.order)

# Use Auto SARIMA to find the best (p,d,q)(P,D,Q,s) for SARIMA
sarima_model = pm.auto_arima(train_data['Sales'], seasonal=True, m=7, 
                             trace=True, suppress_warnings=True, stepwise=True)

# Print best parameters
print("Best SARIMA Parameters:", sarima_model.order, sarima_model.seasonal_order)
"""
```

```python

# ARIMA Model Training
arima_model = ARIMA(train_data['Sales'], order=(5,1,0))  # (p,d,q) needs tuning
arima_model_fit = arima_model.fit()

# SARIMA Model Training
sarima_model = SARIMAX(train_data['Sales'], 
                       order=(5,1,0), 
                       seasonal_order=(1,1,1,7))  # (P,D,Q,s) for weekly seasonality
sarima_model_fit = sarima_model.fit()


# Step 3: Predict Last 14 Days & Measure Performance
# Predict last 14 days using ARIMA
arima_predictions = arima_model_fit.forecast(steps=14)

# Predict last 14 days using SARIMA
sarima_predictions = sarima_model_fit.forecast(steps=14)

# Convert test data to numpy array for comparison
y_true = test_data['Sales'].values

# Compute performance metrics
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100  # Percentage error
    print(f"{model_name} Performance:")
    print(f"  MAE:  {mae:.2f}")
    print(f"  RMSE: {rmse:.2f}")
    print(f"  MAPE: {mape:.2f}%\n")

# Evaluate ARIMA
evaluate_model(y_true, arima_predictions, "ARIMA")

# Evaluate SARIMA
evaluate_model(y_true, sarima_predictions, "SARIMA")
```

```python

# Step 4: Forecast 14 Days into the Future
# Forecast next 14 days with ARIMA
future_arima_forecast = arima_model_fit.forecast(steps=14)

# Forecast next 14 days with SARIMA
future_sarima_forecast = sarima_model_fit.forecast(steps=14)

# Create future date index
future_dates = pd.date_range(df_full.index[-1] + pd.Timedelta(days=1), periods=14, freq='D')

# Store results in DataFrame
future_forecasts = pd.DataFrame({
    'Date': future_dates,
    'ARIMA_Predicted_Sales': future_arima_forecast,
    'SARIMA_Predicted_Sales': future_sarima_forecast
})


# Step 5: Plot Predictions vs. Actual
plt.figure(figsize=(12,5))
plt.plot(df_full.index[-30:], df_full['Sales'][-30:], marker='o', linestyle='-', label="Actual Sales", color='blue')
plt.plot(test_data.index, arima_predictions, marker='o', linestyle='dashed', label="ARIMA Predictions", color='red')
plt.plot(test_data.index, sarima_predictions, marker='o', linestyle='dashed', label="SARIMA Predictions", color='green')
plt.axvline(x=test_data.index[0], color='black', linestyle='--', label="Start of Test Data")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Actual vs Predicted Sales (ARIMA vs SARIMA)")
plt.legend()
plt.xticks(rotation=45)
plt.show()


# Step 6: Print Future 14-Day Forecast

print("Future 14-Day Sales Forecast:")
print(future_forecasts)
```

* The ARIMA and SARIMA models were used to forecast the last 14 days of sales based on previous sales history
* There is an opportunity to further tune these models but an attempt at auto tuning was made and the results were worse then this manual set of model parameters
* ARIMA and SARIMA fail to capture the highly volatile nature of the sales in these 14 days

# XGBoost
* Used an XGBoost model with additional features (rolling 7 day average, rolling 14 day average, sales lags over the previous two weeks, etc)
* Started off with a random set of hyperparamters
* Did not use any of the categorical features (products, subcategories, regions)

```python

import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# Step 1: Prepare Data for Training & Testing
# Define features (X) and target variable (y)
X = df_full.drop(columns=['Sales'])  # Exclude target variable
y = df_full['Sales']

# Split into train & test sets
train_size = len(df_full) - 14  # Keep last 14 days for testing
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:].copy()
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

# Step 2: Train XGBoost Model
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.05,
    max_depth=5,
    n_estimators=200,
    subsample=0.9,
    random_state=42
)

# Train the model
xgb_model.fit(X_train, y_train)

# Step 3: Recursive Forecasting for 14 Days
# List to store predictions
recursive_preds = []

# Make a copy of X_test for recursive updates
X_test_copy = X_test.copy()

# Predict one day at a time, updating lags dynamically
for i in range(len(X_test_copy)):
    # Predict sales for the current day
    pred = xgb_model.predict(X_test_copy.iloc[i:i+1])[0]
    
    # Store the prediction
    recursive_preds.append(pred)

    # Update lag features dynamically for future predictions
    if i + 1 < len(X_test_copy):  # Prevent out-of-bounds error
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_1')] = pred
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_2')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_1')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_3')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_2')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_4')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_3')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_5')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_4')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_6')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_5')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_7')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_6')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_8')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_7')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_9')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_8')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_10')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_9')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_11')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_10')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_12')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_11')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_13')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_12')]
        X_test_copy.iloc[i + 1, X_test_copy.columns.get_loc('Sales_Lag_14')] = X_test_copy.iloc[i, X_test_copy.columns.get_loc('Sales_Lag_13')]

# Convert list of predictions to NumPy array
xgb_predictions = np.array(recursive_preds)

# Step 4: Evaluate Model Performance
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    
    # Fix for zero values in actual sales to avoid division by zero
    epsilon = 1e-5
    mape = np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100  

    print(f"{model_name} Performance:")
    print(f"  MAE:  {mae:.2f}")
    print(f"  RMSE: {rmse:.2f}")
    print(f"  MAPE: {mape:.2f}%\n")


evaluate_model(y_test, xgb_predictions, "XGBoost (Recursive)")

# Step 5: Plot Results
plt.figure(figsize=(12,5))
plt.plot(y_test.index, y_test, marker='o', linestyle='-', label="Actual Sales", color='blue')
plt.plot(y_test.index, xgb_predictions, marker='o', linestyle='dashed', label="XGBoost Predictions (Recursive)", color='red')
plt.axvline(x=y_test.index[0], color='black', linestyle='--', label="Start of Test Data")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Actual vs Predicted Sales (XGBoost Recursive)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

* if MAPE is what we are targeting or using as a gating metric , then ARIMA/SARIMA are performing better. this may be do to XGBoost overestimating and inflating error on zero sales days

## XGBoost - Log Transformation
* To improve perfromance, a log transformation on the sales data to compresses big numbers (like high sales days) and spread out small numbers, making the data more normal looking and easier for models like linear regression or time series forecasting to handle
* Used a random set of hyperparameters first and then used grid search to optimize hyperparamters but results were similar, no noticeable performance improvements

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Step 1: Load & Preprocess Data
# Load dataset from CSV
df = pd.read_csv("../Dataset.csv")

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")

# Aggregate sales to daily level (if not already done)
df_grouped = df.groupby(['Order Date'])['Sales'].sum().reset_index()

# **Apply Log Transformation to Sales**
df_grouped['Sales'] = np.log1p(df_grouped['Sales'])  # log(1 + Sales)

# **Create Lag Features & Rolling Averages**
for lag in range(1, 15):  # 14-day lag features
    df_grouped[f'Sales_Lag_{lag}'] = df_grouped['Sales'].shift(lag)

df_grouped['Sales_Rolling_Mean_7'] = df_grouped['Sales'].rolling(7).mean()
df_grouped['Sales_Rolling_Mean_14'] = df_grouped['Sales'].rolling(14).mean()

# Drop NaN values (due to lag/rolling calculations)
df_grouped.dropna(inplace=True)

# Set 'Order Date' as the index**
df_grouped.set_index('Order Date', inplace=True)


# Step 2: Train-Test Split
train_size = len(df_grouped) - 14  # Keep last 14 days for testing
train_data = df_grouped.iloc[:train_size]
test_data = df_grouped.iloc[train_size:].copy()

# **Ensure test_data index is in datetime format**
test_data.index = pd.to_datetime(test_data.index)

# Features & Target
X_train, y_train = train_data.drop(columns=['Sales']), train_data['Sales']
X_test, y_test = test_data.drop(columns=['Sales']), test_data['Sales']


# Step 3: Train XGBoost Model
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.05,
    max_depth=5,
    n_estimators=200,
    subsample=0.9,
    random_state=42
)

# Train the model
xgb_model.fit(X_train, y_train)


# Step 4: Recursive Forecasting for 14 Days
# List to store predictions
recursive_preds = []

# **Ensure X_pred starts from the correct test date range**
X_pred = X_test.copy()
forecast_index = test_data.index  # Use actual forecast period

# Predict one day at a time, updating lags dynamically
for i in range(len(X_test)):
    # Predict sales for the current day
    pred = xgb_model.predict(X_pred.iloc[i:i+1])[0]
    recursive_preds.append(pred)

    # Update lag features dynamically for future predictions
    if i + 1 < len(X_test):  # Prevent out-of-bounds error
        X_pred.iloc[i + 1, X_pred.columns.get_loc('Sales_Lag_1')] = pred
        for lag in range(2, 15):  # Shift previous lags
            if i + 1 >= lag:
                X_pred.iloc[i + 1, X_pred.columns.get_loc(f'Sales_Lag_{lag}')] = X_pred.iloc[i + 1 - lag, X_pred.columns.get_loc('Sales_Lag_1')]

# **Reverse Log Transformation**
xgb_predictions = np.expm1(recursive_preds)  # Convert back to original scale
y_test_original = np.expm1(y_test.values)  # Convert actual test sales back


# Step 5: Evaluate Performance

def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))

    # Fix for zero values in actual sales to avoid division by zero
    epsilon = 1e-5
    mape = np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100  

    print(f" {model_name} Performance:")
    print(f"   MAE:  {mae:.2f}")
    print(f"   RMSE: {rmse:.2f}")
    print(f"   MAPE: {mape:.2f}%\n")

evaluate_model(y_test_original, xgb_predictions, "XGBoost (Log-Transformed Recursive)")


# Step 6: Plot Results
fig, ax = plt.subplots(figsize=(12, 6))

# Plot actual and predicted values
ax.plot(forecast_index, y_test_original, 'bo-', label="Actual Sales")
ax.plot(forecast_index, xgb_predictions, 'r--o', label="XGBoost Predictions (Log)")

# Mark the start of test data
ax.axvline(forecast_index[0], color='black', linestyle='dashed', label="Start of Test Data")

# Format X-axis properly with date labels
ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.title("Actual vs Predicted Sales (XGBoost with Log Transformation)")
plt.show()
```

## XGBoost - Log Transformation - Optimized Hyperparamters

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV


# Step 1: Load & Preprocess Data
# Load dataset from CSV
df = pd.read_csv("../Dataset.csv")

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")

# Aggregate sales to daily level (if not already done)
df_grouped = df.groupby(['Order Date'])['Sales'].sum().reset_index()

# **Apply Log Transformation to Sales**
df_grouped['Sales'] = np.log1p(df_grouped['Sales'])  # log(1 + Sales)

# **Create Lag Features & Rolling Averages**
for lag in range(1, 15):  # 14-day lag features
    df_grouped[f'Sales_Lag_{lag}'] = df_grouped['Sales'].shift(lag)

df_grouped['Sales_Rolling_Mean_7'] = df_grouped['Sales'].rolling(7).mean()
df_grouped['Sales_Rolling_Mean_14'] = df_grouped['Sales'].rolling(14).mean()

# Drop NaN values (due to lag/rolling calculations)
df_grouped.dropna(inplace=True)

# **Set 'Order Date' as the index**
df_grouped.set_index('Order Date', inplace=True)


# Step 2: Train-Test Split
train_size = len(df_grouped) - 14  # Keep last 14 days for testing
train_data = df_grouped.iloc[:train_size]
test_data = df_grouped.iloc[train_size:].copy()

# **Ensure test_data index is in datetime format**
test_data.index = pd.to_datetime(test_data.index)

# Features & Target
X_train, y_train = train_data.drop(columns=['Sales']), train_data['Sales']
X_test, y_test = test_data.drop(columns=['Sales']), test_data['Sales']


# Step 3: Hyperparameter Optimization
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 300, 500],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 0.9, 1.0]
}

xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

tscv = TimeSeriesSplit(n_splits=5)  # Time-based cross-validation

grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='neg_mean_absolute_error',  # Minimize error
    cv=tscv,
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

# Best Parameters
best_params = grid_search.best_params_
print(f" Best Hyperparameters: {best_params}")


# Step 4: Train XGBoost with Optimized Hyperparameters
best_xgb = xgb.XGBRegressor(**best_params, objective='reg:squarederror', random_state=42)
best_xgb.fit(X_train, y_train)


# Step 5: Recursive Forecasting for 14 Days
# List to store predictions
recursive_preds = []

# **Ensure X_pred starts from the correct test date range**
X_pred = X_test.copy()
forecast_index = test_data.index  # Use actual forecast period

# Predict one day at a time, updating lags dynamically
for i in range(len(X_test)):
    # Predict sales for the current day
    pred = best_xgb.predict(X_pred.iloc[i:i+1])[0]
    recursive_preds.append(pred)

    # Update lag features dynamically for future predictions
    if i + 1 < len(X_test):  # Prevent out-of-bounds error
        X_pred.iloc[i + 1, X_pred.columns.get_loc('Sales_Lag_1')] = pred
        for lag in range(2, 15):  # Shift previous lags
            if i + 1 >= lag:
                X_pred.iloc[i + 1, X_pred.columns.get_loc(f'Sales_Lag_{lag}')] = X_pred.iloc[i + 1 - lag, X_pred.columns.get_loc('Sales_Lag_1')]

# **Reverse Log Transformation**
xgb_predictions = np.expm1(recursive_preds)  # Convert back to original scale
y_test_original = np.expm1(y_test.values)  # Convert actual test sales back


# Step 6: Evaluate Performance
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))

    # Fix for zero values in actual sales to avoid division by zero
    epsilon = 1e-5
    mape = np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100  

    print(f" {model_name} Performance:")
    print(f"   MAE:  {mae:.2f}")
    print(f"   RMSE: {rmse:.2f}")
    print(f"   MAPE: {mape:.2f}%\n")

evaluate_model(y_test_original, xgb_predictions, "Optimized XGBoost (Log-Transformed Recursive)")


# Step 7: Plot Results
fig, ax = plt.subplots(figsize=(12, 6))

# Plot actual and predicted values
ax.plot(forecast_index, y_test_original, 'bo-', label="Actual Sales")
ax.plot(forecast_index, xgb_predictions, 'r--o', label="Optimized XGBoost Predictions (Log)")

# Mark the start of test data
ax.axvline(forecast_index[0], color='black', linestyle='dashed', label="Start of Test Data")

# Format X-axis properly with date labels
ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.title("Actual vs Predicted Sales (Optimized XGBoost with Log Transformation)")
plt.show()
```

## Notes
Auto regregressive integration moving average

- the model trained in a sense, the model parameters are optimized to find the best set of parameters 

AR order - p = how many past observations are considered for the prediction (previous days sales)
I order d = how many differences are necessary to make the data stationary (looking for trends in sales long term and removing them to make it stationary) 
MA order - q = considers past days errors to smooth out forecasts 

- provide the model the most recent data necessary for the model (based on p and q)

- stationary - mean / std dev / variance is consistent statistical mean / std dev / variance over time

- performance metrics

- RMSE / MAE / MAPE

- Limitations

- non stationary (differencing might not be enough) / multiple seasons (daily patterns / monthly patterns / yearly patterns) / non-linear relationships / needs a lot of data data / can not handle multivariate data / major changes (new product launches)

- Great for simpler stationary time series data - can be used as inputs for non-linear models that rely on othe paramters that can be estimated with ARIMA


SARIMA 
 - great for capturing seasons
 - limitations
 - non-stationary / struggles with non-linear / needs a lot of data / can not handle multivariate data / major changes (market disruptions)


- if data is too complex then other models can be explored to improve predictive performance
- RMSE/MAE/MAPE - focus on a mape of sub 10%
- optimize parameters with auto tuning or grid search
- test for stationarity - augmented dickey fuller test
- options (log transformation may not be necessary if the data is already stationary)
- 1 - using arima with optimized parameters
- 2 - using arima with log transformed sales and optimized parameters
- 3 - using sarima with optimized parameters
- 4 - using sarima with log transfotmed sales and optimized parameters
- can these capture data patterns effectively?

VAR - vector autoregressive (not going to try for now)
- assumes linearity and stationarity

## Explore using categorical features 

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV


# Step 1: Load & Preprocess Data
# Load dataset from CSV
df = pd.read_csv("../Dataset.csv")

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")

# Group by date and sub-category to get daily sales per sub-category
subcategory_daily_sales = df.groupby(['Order Date', 'Sub-Category'])['Sales'].sum().reset_index()

# Group by date to get total daily sales
total_daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index().rename(columns={'Sales': 'Total_Sales'})

# Merge to calculate share
subcategory_daily_sales = pd.merge(subcategory_daily_sales, total_daily_sales, on='Order Date')

# Calculate the share
subcategory_daily_sales['Sales_Share'] = subcategory_daily_sales['Sales'] / subcategory_daily_sales['Total_Sales']
```

```python
import matplotlib.pyplot as plt

# Select the subcategories you want to inspect
selected_subcategories = ['Copiers', 'Machines']

plt.figure(figsize=(14, 6))

for subcat in selected_subcategories:
    subcat_data = subcategory_daily_sales[subcategory_daily_sales['Sub-Category'] == subcat]
    plt.plot(subcat_data['Order Date'], subcat_data['Sales_Share'], label=subcat)

plt.xlabel('Date')
plt.ylabel('Sales Share')
plt.title('Daily Sales Share per Subcategory')
plt.legend()
plt.show()
```

```python
# Variability summary
share_variability = subcategory_daily_sales.groupby('Sub-Category')['Sales_Share'].agg(['mean', 'std', 'var']).reset_index()

print(share_variability)
```

```python
subcategory_daily_sales['WeekOfYear'] = subcategory_daily_sales['Order Date'].dt.isocalendar().week

# Average share per week for each sub-category
weekly_avg_share = subcategory_daily_sales.groupby(['Sub-Category', 'WeekOfYear'])['Sales_Share'].mean().reset_index()

# Plot
plt.figure(figsize=(14, 6))
for subcat in selected_subcategories:
    subcat_data = weekly_avg_share[weekly_avg_share['Sub-Category'] == subcat]
    plt.plot(subcat_data['WeekOfYear'], subcat_data['Sales_Share'], label=subcat)

plt.xlabel('Week of Year')
plt.ylabel('Average Sales Share')
plt.title('Seasonality Check: Average Sales Share by Week of Year')
plt.legend()
plt.show()
```

## XGBoost with XGBoost
### Leverage XGBoost to make predictions on subcategory sales numbers that can be used for predictions

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV


# Utility Functions
def optimize_xgboost(X_train, y_train, param_grid, n_splits=5):
    """Run Grid Search to optimize XGBoost hyperparameters."""
    model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
    tscv = TimeSeriesSplit(n_splits=n_splits)
    grid_search = GridSearchCV(estimator=model,
                               param_grid=param_grid,
                               scoring='neg_mean_absolute_error',
                               cv=tscv,
                               verbose=1,
                               n_jobs=-1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_, grid_search.best_params_



def evaluate_model(y_true, y_pred, model_name):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))

    # Mask to avoid division by zero (for MAPE)
    mask = y_true > 1e-5

    if np.sum(mask) == 0:
        mape = np.nan
    else:
        mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

    print(f"\n {model_name} Performance:")
    print(f"   MAE:  {mae:.5f}")
    print(f"   RMSE: {rmse:.5f}")
    
    if not np.isnan(mape):
        print(f"   MAPE: {mape:.2f}%\n")
    else:
        print(f"   MAPE: Not computed due to zero denominators.\n")



# Step 1: Load & Preprocess Data
df = pd.read_csv('../Dataset.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")

# Aggregate sales by date & sub-category
df_daily = df.groupby(['Order Date', 'Sub-Category'])['Sales'].sum().reset_index()

# Pivot to get total sales and sub-category sales per day
df_total = df_daily.groupby('Order Date')['Sales'].sum().rename('Total_Sales').reset_index()
df_copiers = df_daily[df_daily['Sub-Category'] == 'Copiers'].groupby('Order Date')['Sales'].sum().rename('Copiers_Sales').reset_index()
df_machines = df_daily[df_daily['Sub-Category'] == 'Machines'].groupby('Order Date')['Sales'].sum().rename('Machines_Sales').reset_index()

# Merge and fill missing dates with 0
df_merged = pd.merge(df_total, df_copiers, on='Order Date', how='left').fillna(0)
df_merged = pd.merge(df_merged, df_machines, on='Order Date', how='left').fillna(0)

# Compute sales shares
df_merged['Copiers_Share'] = df_merged['Copiers_Sales'] / df_merged['Total_Sales']
df_merged['Machines_Share'] = df_merged['Machines_Sales'] / df_merged['Total_Sales']


# Step 2: Feature Engineering for Shares
df_features = df_merged.copy()

# Add time-based features
df_features['Day_of_Week'] = df_features['Order Date'].dt.dayofweek
df_features['Week_of_Year'] = df_features['Order Date'].dt.isocalendar().week.astype(int)
df_features['Month'] = df_features['Order Date'].dt.month

# Lag features for shares
for lag in range(1, 15):
    df_features[f'Copiers_Share_Lag_{lag}'] = df_features['Copiers_Share'].shift(lag)
    df_features[f'Machines_Share_Lag_{lag}'] = df_features['Machines_Share'].shift(lag)

# Drop initial NaNs
df_features.dropna(inplace=True)
df_features.set_index('Order Date', inplace=True)


# Step 3: Train-Test Split for Share Models
train_size = len(df_features) - 14
train_data = df_features.iloc[:train_size]
test_data = df_features.iloc[train_size:]

# Features and Targets for Share Models
X_train_share = train_data.drop(columns=['Total_Sales', 'Copiers_Sales', 'Machines_Sales', 'Copiers_Share', 'Machines_Share'])
y_train_copiers_share = train_data['Copiers_Share']
y_train_machines_share = train_data['Machines_Share']

X_test_share = test_data.drop(columns=['Total_Sales', 'Copiers_Sales', 'Machines_Sales', 'Copiers_Share', 'Machines_Share'])


# Step 4: Optimize and Train Share Models
param_grid = {
    'learning_rate': [0.05],
    'n_estimators': [300],
    'max_depth': [5],
    'subsample': [0.9]
}

copiers_model, _ = optimize_xgboost(X_train_share, y_train_copiers_share, param_grid)
machines_model, _ = optimize_xgboost(X_train_share, y_train_machines_share, param_grid)


# Step 5: Recursive Prediction of Shares
recursive_copiers_share_preds = []
recursive_machines_share_preds = []

X_pred_share = X_test_share.copy()

for i in range(len(X_test_share)):
    pred_copiers = copiers_model.predict(X_pred_share.iloc[i:i+1])[0]
    pred_machines = machines_model.predict(X_pred_share.iloc[i:i+1])[0]
    
    recursive_copiers_share_preds.append(pred_copiers)
    recursive_machines_share_preds.append(pred_machines)
    
    # Update lag values recursively
    if i + 1 < len(X_test_share):
        X_pred_share.iloc[i + 1, X_pred_share.columns.get_loc('Copiers_Share_Lag_1')] = pred_copiers
        X_pred_share.iloc[i + 1, X_pred_share.columns.get_loc('Machines_Share_Lag_1')] = pred_machines
        
        for lag in range(2, 15):
            if i + 1 >= lag:
                X_pred_share.iloc[i + 1, X_pred_share.columns.get_loc(f'Copiers_Share_Lag_{lag}')] = X_pred_share.iloc[i + 1 - lag, X_pred_share.columns.get_loc('Copiers_Share_Lag_1')]
                X_pred_share.iloc[i + 1, X_pred_share.columns.get_loc(f'Machines_Share_Lag_{lag}')] = X_pred_share.iloc[i + 1 - lag, X_pred_share.columns.get_loc('Machines_Share_Lag_1')]

# Evaluate Share Models
evaluate_model(test_data['Copiers_Share'], recursive_copiers_share_preds, "Copiers Share Prediction")
evaluate_model(test_data['Machines_Share'], recursive_machines_share_preds, "Machines Share Prediction")

# Step 6: Train & Predict Total Sales Model


# Prepare data for global model
df_sales = df_merged.copy()

# Add lag features and rolling means for total sales
df_sales['Sales_Log'] = np.log1p(df_sales['Total_Sales'])

for lag in range(1, 15):
    df_sales[f'Sales_Lag_{lag}'] = df_sales['Sales_Log'].shift(lag)

df_sales['Sales_Rolling_Mean_7'] = df_sales['Sales_Log'].rolling(7).mean()
df_sales['Sales_Rolling_Mean_14'] = df_sales['Sales_Log'].rolling(14).mean()

# Drop NaNs
df_sales.dropna(inplace=True)
df_sales.set_index('Order Date', inplace=True)

# Merge predicted shares for test data
df_sales.loc[test_data.index, 'Copiers_Share'] = recursive_copiers_share_preds
df_sales.loc[test_data.index, 'Machines_Share'] = recursive_machines_share_preds

# Train/Test split
train_size = len(df_sales) - 14
train_global = df_sales.iloc[:train_size]
test_global = df_sales.iloc[train_size:]

X_train_global = train_global.drop(columns=['Total_Sales', 'Copiers_Sales', 'Machines_Sales', 'Sales_Log'])
y_train_global = train_global['Sales_Log']

X_test_global = test_global.drop(columns=['Total_Sales', 'Copiers_Sales', 'Machines_Sales', 'Sales_Log'])
y_test_global = test_global['Sales_Log']

global_features = X_train_global.columns.tolist()

# Optimize and Train Global Model
global_model, _ = optimize_xgboost(X_train_global, y_train_global, param_grid)


# Step 7: Recursive Forecasting of Total Sales
recursive_sales_preds = []
X_pred_global = X_test_global[global_features].copy()


for i in range(len(X_test_global)):
    pred_log_sales = global_model.predict(X_pred_global.iloc[i:i+1])[0]
    recursive_sales_preds.append(pred_log_sales)
    
    if i + 1 < len(X_test_global):
        X_pred_global.iloc[i + 1, X_pred_global.columns.get_loc('Sales_Lag_1')] = pred_log_sales
        for lag in range(2, 15):
            if i + 1 >= lag:
                X_pred_global.iloc[i + 1, X_pred_global.columns.get_loc(f'Sales_Lag_{lag}')] = X_pred_global.iloc[i + 1 - lag, X_pred_global.columns.get_loc('Sales_Lag_1')]

# Reverse log transform
final_sales_preds = np.expm1(recursive_sales_preds)
y_test_actual = test_global['Total_Sales'].values


# Step 8: Evaluate Total Sales Model
evaluate_model(y_test_actual, final_sales_preds, "Total Sales Prediction (Recursive)")


# Step 9: Plot Results

forecast_index = test_global.index

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(forecast_index, y_test_actual, 'bo-', label="Actual Sales")
ax.plot(forecast_index, final_sales_preds, 'r--o', label="XGBoost Total Sales Prediction")

ax.axvline(forecast_index[0], color='black', linestyle='dashed', label="Start of Test Data")
ax.xaxis.set_major_locator(mdates.DayLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Actual vs Predicted Total Sales (Recursive with Shares)")
plt.legend()
plt.show()
```
